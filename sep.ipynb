```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import librosa
import os

import librosa.display
from PIL import Image
from IPython.display import Audio, display

from pydub import AudioSegment

from tqdm import tqdm
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("test.tsv",  sep='\t') 
df = df[df['gender'].notna()]
```

```python
df_male = df[df['gender'] == 'male']
df_female = df[df['gender'] == 'female']
```

# Extract Data

```python
WORD_LENGTH = 5
TIME_DURATION = [5.0, 5.5]
SR = 16000
```

```python
males_file = []

for index, row in df_male.iterrows():
    print(index)
    audio, sr = librosa.load('clips/' + row['path'], sr = SR)
    sentence_char = len(row['sentence'].split(" "))
    
    if(sentence_char < WORD_LENGTH):
        continue
    
    duration = librosa.get_duration(audio, sr)
    
    if(duration >= TIME_DURATION[0] and duration <= TIME_DURATION[1]):
        males_file.append('clips/' + row['path'])
    
    if(len(males_file) == 100): break
```

```python
pd.DataFrame({'path':males_file}).to_csv('male_100.csv')
```

```python
female_file = []

for index, row in df_female.iterrows():
    audio, sr = librosa.load('clips/' + row['path'], sr = SR)
    sentence_char = len(row['sentence'].split(" "))
    
    if(sentence_char < WORD_LENGTH):
        continue
    
    duration = librosa.get_duration(audio, sr)
    
    if(duration >= TIME_DURATION[0] and duration <= TIME_DURATION[1]):
        female_file.append('clips/' + row['path'])
    
    if(len(female_file) == 100): break
    else: print(len(female_file))
```

```python
pd.DataFrame({'path':female_file}).to_csv('female_100.csv')
```

```python
len(female_file)
```

# To WAV

```python
for path in tqdm(males_file):
    audio = AudioSegment.from_mp3(path)
    name = path.split('/')[1].split('.')[0]
    audio.export("male_100/"+ name + ".wav", format="wav")
```

```python
for path in tqdm(female_file):
    audio = AudioSegment.from_mp3(path)
    name = path.split('/')[1].split('.')[0]
    audio.export("female_100/"+ name + ".wav", format="wav")
```

## To WAV

```python
ages = [20, 30, 40, 50, 60]
sex = 'male'
```

```python
for age in tqdm(ages):
    df = pd.read_csv(sex + "_50_age_" + str(age) + "_train.csv")
    paths = list(df['path'])
    os.system('mkdir ' + sex + "_50_age_" + str(age) + "_train")
    for path in paths:
        audio = AudioSegment.from_mp3(path)
        name = path.split('/')[1].split('.')[0]
        audio.export(sex + "_50_age_" + str(age) + "_train" + "/"+ name + ".wav", format="wav")
```
